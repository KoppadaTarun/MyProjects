```python
from io import StringIO

data = ''' Name,Age,Income($)
Rob,27,70000
Michael,29,90000
Mohan,29,61000
Ismail,28,60000
Kory,42,150000
Gautam,39,155000
David,41,160000
Andrea,38,162000
Brad,36,156000
Angelina,35,130000
Donald,37,137000
Tom,26,45000
Arnold,27,48000
Jared,28,51000
Stark,29,49500
Ranbir,32,53000
Dipika,40,65000
Priyanka,41,63000
Nick,43,64000
Alia,39,80000
Sid,41,82000
Abdul,39,58000
'''
```

```python
data = StringIO(data)
```

```python
import pandas as pd
df = pd.read_csv(data)
df
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.xlabel("Age")
plt.ylabel("Income $")
plt.scatter(df.Age, df["Income($)"], color= "blue", marker="+")
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[["Age","Income($)"]])
y_predicted
```

```python
df['cluster'] = y_predicted
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1["Age"], df1["Income($)"])
plt.scatter(df2["Age"], df2["Income($)"])
plt.scatter(df3["Age"], df3["Income($)"])
```

```python
scaler = MinMaxScaler()
scaler.fit(df[["Income($)"]])
df["Income($)"] = scaler.transform(df[["Income($)"]])
scaler.fit(df[["Age"]])
df["Age"] = scaler.transform(df[["Age"]])
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1["Age"], df1["Income($)"])
plt.scatter(df2["Age"], df2["Income($)"])
plt.scatter(df3["Age"], df3["Income($)"])
```

```python
km = KMeans(n_clusters=3)
y_predict = km.fit_predict(df[["Age","Income($)"]])
```

```python
y_predict
```

```python
df.drop('cluster', axis="columns")
df['cluster'] = y_predict
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1["Age"], df1["Income($)"], color="green")
plt.scatter(df2["Age"], df2["Income($)"], color="red")
plt.scatter(df3["Age"], df3["Income($)"], color="black")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], color="purple", marker="*",label="centroid")
plt.legend()
```

```python
km.cluster_centers_
```

```python
k_rang = range(1,10)
sse= []
for k in k_rang:
  km = KMeans(n_clusters=k)
  km.fit_predict(df[['Age','Income($)']])
  sse.append(km.inertia_)
print(sse)
```

```python
plt.xlabel("k")
plt.ylabel("SSE")
plt.plot(k_rang, sse)
```
